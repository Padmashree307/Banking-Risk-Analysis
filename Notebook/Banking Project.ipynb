## Heatmaps

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'red'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

## Insights of EDA:

1. Income Band is a Strong Risk Indicator

Customers in the Low Income Band are more frequently associated with higher risk categories.
Lower income limits a customer’s ability to repay loans, making them more prone to default. Segmenting income levels helps banks prioritize whom to assess more strictly during credit evaluation.

2. High Credit Card Balances Correlate with Bank Loans

Customers with high credit card balances also tend to have larger outstanding bank loans.
This pattern suggests financial overextension. High unsecured debt (like credit cards) combined with existing loans increases the probability of payment difficulties, signaling higher risk.

3. Loyalty Classification Impacts Risk Weighting

Customers with lower loyalty classification are more likely to have higher risk weights.
Loyalty classification may reflect customer history, tenure, or activity level. Short-term or disengaged customers often lack repayment history, making their creditworthiness less predictable.

4. Savings & Checking Accounts Reduce Default Risk  

Presence of active savings or checking accounts is common among lower-risk customers.
These accounts indicate stable financial behavior and provide liquidity cushions, which reduce the chance of payment delays or defaults. They are often used in credit scoring as indicators of financial responsibility.

5. Superannuation and Deposits Reflect Financial Strength 

Higher superannuation savings and bank deposits are observed in more financially stable profiles.
These assets act as financial buffers. Customers with higher reserves are less likely to default, especially in economic downturns, making them safer lending candidates.
